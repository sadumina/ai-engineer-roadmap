```python
import pandas as pd
import numpy as np 
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
raw_data = pd.read_csv("train.csv")
```

```python
raw_data["Age"].hist(bins=30)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("distribution of passengers")
plt.show()
```

```python
sns.countplot(data=raw_data,x="Sex",hue="Survived")
```

```python
sns.barplot(data=raw_data,x="Pclass" , y="Survived")
```

```python
sns.histplot(data=raw_data,x="Fare",hue="Survived",bins=40,kde=True)
```

```python
sns.countplot(data=raw_data, x="Embarked", hue="Survived")
plt.title("Survival Count by Port of Embarkation")
plt.show()
```
