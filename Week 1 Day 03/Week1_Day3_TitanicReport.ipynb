```python
import pandas as pd
```

```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#multiple plots
raw_data = pd.read_csv("train.csv")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,5)) 
raw_data["Age"].hist(bins=30 ,ax=axes[0] ,color="skyblue")
axes[0].set_title("Age distribution")

raw_data["Fare"].hist(bins=30,ax=axes[1],color="orange") 
axes[1].set_title("Fare distribution")

plt.tight_layout() 
plt.show()
```

```python
#Correlation Heatmaps
plt.figure(figsize=(10,6))
numeric_data = raw_data.select_dtypes(include=["int64", "float64"])
sns.heatmap(numeric_data.corr(),annot=True ,cmap="coolwarm",fmt=".2f")
plt.title("correlation Heatmap")
plt.show()
```

```python
#Gender + Class Survival (Combined View)
sns.catplot(data=raw_data,x="Sex",y="Survived",hue="Pclass",kind ="bar")
plt.title("Survival Rate by Gender & Passenger Class")
plt.show()
```

```python
### Insight: Age & Fare Distributions
Most passengers were between 2040 years old.  
Higher ticket fares were rare but linked to 1st class passengers.  

### Insight: Correlation Heatmap
Survival is positively correlated with being female and higher class.  
Age had little correlation with survival.  
```
